```python
# import important libraries 
import os
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# file paths
file_path_stardist = r"/Users/Prags/Desktop/Stardist_analysis /YES_Fixed_091024_labels/Results_003.csv"
file_path_fiji = r"/Users/Prags/Desktop/S. pombe image data/241009_AV1200_YES_Fixed/241009 AV1200 YES Pk Fixed/Tif_Images/Results_Tables/60x_60pc_100ms_zstack003_Results.csv"

# Read the CSV files into DataFrames
df_stardist = pd.read_csv(file_path_stardist)
df_fiji = pd.read_csv(file_path_fiji)

```

```python
bin_width = 0.3  # Desired bin width
# Create bins with the specified width
stardist_bins = np.arange(min(df_stardist["Area"]), max(df_stardist["Area"]) + bin_width, bin_width)
fiji_bins = np.arange(min(df_fiji["Area"]), max(df_fiji["Area"]) + bin_width, bin_width)


plt.hist(df_stardist["Area"], bins=stardist_bins, color='g', alpha=0.7)
plt.hist(df_fiji["Area"], bins=fiji_bins, color='b', alpha=0.5)


# Add labels and title
plt.xlabel('Area of Nucleus (microns$^2$)')
plt.ylabel('Frequency')
plt.title('Area of Different Nuclei Stardist vs MacroJ')
plt.legend()
```
