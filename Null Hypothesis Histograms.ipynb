# Histograms 

# Null Hypothesis One 
There is no significant difference in cell growth/nuclei size when the cells are grown in YES and EMM 
- Fixed vs Fixed 
- Live vs Live

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
results_EMM = r"/Users/Prags/Desktop/S. pombe image data/301024 AV1200 EMM Fixed/Tif_Images/Results_Tables/EMM_Fixed_all.csv"
results_YES = r"/Users/Prags/Desktop/S. pombe image data/241009_AV1200_YES_Fixed/YES_Fixed_All.csv"

df_EMM = pd.read_csv(results_EMM)
df_YES = pd.read_csv(results_YES)

bin_width = 0.3  # Desired bin width
# Create bins with the specified width
EMM_bins = np.arange(min(df_EMM["Area"]), max(df_EMM["Area"]) + bin_width, bin_width)
YES_bins = np.arange(min(df_YES["Area"]), max(df_YES["Area"]) + bin_width, bin_width)


plt.hist(df_EMM["Area"], bins=EMM_bins, color='g', alpha=0.7, label = "EMM")
plt.hist(df_YES["Area"], bins=YES_bins, color='b', alpha=0.5, label = "YES")

plt.xlabel('Area of Nucleus (microns$^2$)') 
plt.ylabel('Frequency')
plt.title('Area of Different Nuclei in EMM vs YES in fixed conditions') 
plt.legend(title="Colour Key")

plt.show()
```

```python
df_EMM.describe()
```

```python
df_YES.describe()
```

#### T-test result:

T-statistic: 12.411851785123899

P-value: 4.010106392427886e-35

the differences in the groups is statistically significant

YES Live vs EMM Live 

```python
results_YES_live = r"/Users/Prags/Desktop/S. pombe image data/241009_AV1200_YES_Live/Tif_Images/Results_Tables/YES_Live_all.csv"
results_EMM_live = r"/Users/Prags/Desktop/S. pombe image data/161024_AV1200_EMM_Live/Tif_Images/Results_Tables/EMM_Live_All.csv"

df_EMM_live = pd.read_csv(results_EMM)
df_YES_live = pd.read_csv(results_YES)

bin_width = 0.3  # Desired bin width
# Create bins with the specified width
EMM_live_bins = np.arange(min(df_EMM_live["Area"]), max(df_EMM_live["Area"]) + bin_width, bin_width)
YES_live_bins = np.arange(min(df_YES_live["Area"]), max(df_YES_live["Area"]) + bin_width, bin_width)


plt.hist(df_EMM_live["Area"], bins=EMM_live_bins, color='g', alpha=0.7, label = "EMM", density = True)
plt.hist(df_YES_live["Area"], bins=YES_live_bins, color='b', alpha=0.5, label = "YES", density = True)

plt.xlabel('Area of Nucleus (microns$^2$)') 
plt.ylabel('Probability Density')  
plt.title('Area of Different Nuclei in EMM vs YES in live imaging') 
plt.legend(title="Colour Key")

plt.show()
```

# Null Hypothesis Three

# Null Hypothesis Four 
there is no significant differance in the data obtained at a low intensity vs at a high intensity

histograms will compare area data found at low intensity vs at a high intensity of cells in the same condtion and in the same ROI/plane of view

```python
# TESTING FOR EMM and EMM-N Fixed - showing the difference between 2 and 3 beacuse visually that is the most significant
results_low = r"/Users/Prags/Desktop/S. pombe image data/131124 AV1200 EMM Fixed Intensity Series/Tif_Images/Results_Tables/All_Channel2.csv"
results_high = r"/Users/Prags/Desktop/S. pombe image data/131124 AV1200 EMM Fixed Intensity Series/Tif_Images/Results_Tables/All_Channel3.csv"
df_low = pd.read_csv(results_low)
df_high = pd.read_csv(results_high)

bin_width = 0.5 # Desired bin width
# Create bins with the specified width
low_bins = np.arange(0.5, 20 + bin_width, bin_width)
high_bins = np.arange(0.5, 20 + bin_width, bin_width)


plt.hist(df_low["Area"], bins=low_bins, color='g', alpha=0.7, label = "Low Intensity")
plt.hist(df_high["Area"], bins=high_bins, color='b', alpha=0.5, label = "High Intensity")

plt.xlabel('Area of Nucleus (microns$^2$)') 
plt.ylabel('Frequency (log scale)')  
plt.yscale('log')
plt.title('Area of The Same Nuclei in Low Intensity (16pc) vs High Intenisty (30pc)') 
plt.legend(title="Colour Key")

plt.show()
```
