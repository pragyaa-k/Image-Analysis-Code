```python
# data for YES fixed and YES live on one histogram 
# to see difference in distribution 

import os
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# give the path for the file that you want to analyse 
table_path1 = r"/Users/Prags/Desktop/241009 AV1200 YES Pk Fixed/Tif_Images/YES_Fixed_PK_All.csv"
table_path2 = r"/Users/Prags/Desktop/241009_AV1200_YES_Live/Tif_Images/YES_Live_all.csv"

# read the files that the path is provided for and store in pandas variable 
df1 = pd.read_csv(table_path1)
df2 = pd.read_csv(table_path2)

# Plot histograms for the "Area" column from both dataframes
plt.hist(df1["Area"], bins=30, color='g', alpha=0.6, label="Fixed PK")
plt.hist(df2["Area"], bins=30, color='b', alpha=0.7, label="Live")

# Add labels and title
plt.xlabel('Area of Nucleus (microns$^2$)')
plt.ylabel('Frequency')
plt.title('Area of Different Nuclei')
plt.legend()

# Show the plot
plt.show()
```
